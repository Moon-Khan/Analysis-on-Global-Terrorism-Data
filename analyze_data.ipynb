### Data Analysis through Visualization

#### Number of Attacks in a Year

```python
years = df['Year'].unique()
no_of_attack = df['Year'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
sns.barplot(x=years, y=no_of_attack, palette='viridis')
plt.title('Number of Attacks in a Year')
plt.xticks(rotation = 90)
plt.xlabel('Years')
plt.ylabel('No of Attacks')
plt.show()
```

#### No of Attacks in Each Year by Region

```python
no_of_attacks_by_region = df.groupby(['Year', 'Region']).size().unstack(fill_value=0)

plt.figure(figsize=(14, 8))
sns.heatmap(no_of_attacks_by_region, cmap='viridis', annot=True, fmt='d', linewidths=.5, cbar_kws={'label': 'Number of Attacks'})
plt.title('Number of Attacks in Each Year by Region')
plt.xlabel('Region')
plt.ylabel('Year')
plt.xticks(rotation=90)
plt.show()
```

### Top 10 Worst Terror Attacks on Countries (2001 and later)

```python
df['Casualties'] = df['Killed'] + df['Wounded']

filtered_df = df[df['Year'] > 2000]

year_country_casualties = filtered_df.groupby(['Year', 'Country'])['Casualties'].sum().reset_index()

top_10_attacks = year_country_casualties.groupby('Year').apply(lambda x: x.nlargest(10, 'Casualties')).reset_index(drop=True)

heatmap_data = top_10_attacks.pivot(index='Year', columns='Country', values='Casualties').fillna(0)

plt.figure(figsize=(14, 8))
sns.heatmap(heatmap_data, cmap='viridis',  fmt='g', linewidths=.5, cbar_kws={'label': 'Number of Casualties'})
plt.title('Top 10 Worst Terror Attacks on Countries (2001 and later)')
plt.xlabel('Country')
plt.ylabel('Year')
plt.yticks(rotation=0)
plt.show()
```

### Top 10 Most Affected Countries by Terror Attacks

```python
df['Casualties'] = df['Killed'] + df['Wounded']

country_casualties = df.groupby('Country')['Casualties'].sum().reset_index()

sorted_countries = country_casualties.sort_values(by='Casualties', ascending=False)

top_10_countries = sorted_countries.head(10)

# Plotting the bar plot
plt.figure(figsize=(12, 6))
plt.bar(top_10_countries['Country'], top_10_countries['Casualties'], color='blue', alpha=0.7)
plt.title('Top 10 Most Affected Countries by Terror Attacks')
plt.xlabel('Country')
plt.ylabel('Number of Casualties')
plt.xticks(rotation=45)
plt.show()
```

### Distribution of Attack Types in Top 10 Most Affected Countries

```python
df['Casualties'] = df['Killed'] + df['Wounded']

country_attack_types = df.groupby(['Country', 'AttackType']).size().unstack(fill_value=0)

top_10_countries = df.groupby('Country')['Casualties'].sum().nlargest(10).index
country_attack_types_top_10 = country_attack_types.loc[top_10_countries]

plt.figure(figsize=(14, 8))
country_attack_types_top_10.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Distribution of Attack Types in Top 10 Most Affected Countries')
plt.xlabel('Country')
plt.ylabel('Number of Attacks')
plt.legend(title='Attack Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

### Top 10 Countries with the Highest Casualties by Region

```python
df['Casualties'] = df['Killed'] + df['Wounded']

country_casualties = df.groupby(['Country', 'Region'])['Casualties'].sum().reset_index()

top_countries = country_casualties.groupby('Country')['Casualties'].sum().nlargest(10).index
top_countries_data = country_casualties[country_casualties['Country'].isin(top_countries)]

plt.figure(figsize=(14, 8))
sns.barplot(x='Casualties', y='Country', hue='Region', data=top_countries_data, palette='viridis')
plt.title('Top 10 Countries with the Highest Casualties by Region')
plt.xlabel('Number of Casualties')
plt.ylabel('Country')
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

### Distribution of Attack Targets

```python
df['Casualties'] = df['Killed'] + df['Wounded']

target_types = df.groupby('Target_type')['Casualties'].sum().reset_index()

plt.figure(figsize=(14, 8))
sns.barplot(x='Casualties', y='Target_type', data=target_types, palette='viridis')
plt.title('Distribution of Attack Targets')
plt.xlabel('Number of Casualties')
plt.ylabel('Target Type')
plt.show()
```

### Distribution of Weapons Used in Attacks

```python
df['Casualties'] = df['Killed'] + df['Wounded']

weapon_types = df.groupby('Weapon_type')['Casualties'].sum().reset_index()

plt.figure(figsize=(14, 8))
sns.barplot(x='Casualties', y='Weapon_type', data=weapon_types, palette='viridis')
plt.title('Distribution of Weapons Used in Attacks')
plt.xlabel('Number of Casualties')
plt.ylabel('Weapon Type')
plt.show()
```
